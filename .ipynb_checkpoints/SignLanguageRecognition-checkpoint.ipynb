```python
# %load './n.py'
import cv2
import mediapipe as mp 
import joblib
import numpy as np 
from gtts import gTTS
import os 
from functions import *
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
s=''
i=0
k=0
current=0
vowels=["ఆ","ఇ","ఉ","ఎ","ఏ", "ఐ", "ఒ", "ఔ", "అం", "అ" ]
dictionary={
  "a":'అ',
  "aa":"ఆ",
  "i":'ఇ',
  "u": "ఉ",
  "e":"ఎ",
  "ai":"ఐ",
  "o": "ఒ",
  "au": "ఔ",
  "am": "అం",
  "ka": "క",
  "ga": "గ",
  "ja": "జ",
  "tta": "ట",
  "da":"డ",
  "ta": "త",
  "dha": "ద",
  "na": "న",
  "pa":"ప",
  "fa":"ఫ",
  "ba": "బ",
  "ma": "మ",
  "ya": "య",
  "ra": "ర",
  "la": "ల",
  "va": "వ",
  "sa": "స",
  "ha":"హ",
  "cha": "చ",
}
hands = mp_hands.Hands(
    min_detection_confidence=0.7, min_tracking_confidence=0.5)

cap = cv2.VideoCapture(0)

def data_clean(landmark):
    data = landmark[0]
    try:
        data = str(data)
        data = data.strip().split('\n')
        garbage = ['landmark {', '  visibility: 0.0', '  presence: 0.0', '}']
        without_garbage = []
        
        for i in data:
            if i not in garbage:
                without_garbage.append(i)
        
        clean = []
        
        for i in without_garbage:
            i = i.strip()
            clean.append(i[2:])
        
        for i in range(0, len(clean)):
            clean[i] = float(clean[i])
            
        return([clean])
    except:
        return(np.zeros([1,63], dtype=int)[0])
print("అక్షరాని చూపెటండి: ")
while cap.isOpened():
    key = cv2.waitKey(1)
    if key==ord('n'):
        print("దాటవేశారు")
        k=(k+1)%3
        if k==0:
            print("అక్షరాని చూపెటండి: ")
        if k==1:
            print("వత్తు చూపెటండి: ")
        if k==2:
            print("గునితాని చూపెటండి: ")
    success, image = cap.read()
    image = cv2.flip(image, 1)
    
    if not success:
        break
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  
    image.flags.writeable = False
    results = hands.process(image)
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
        cleaned_landmark = data_clean(results.multi_hand_landmarks)
#         print(k)
        if cleaned_landmark:
            clf = joblib.load('model.pkl')
            y_pred = clf.predict(cleaned_landmark)
            if i==0:
                current=y_pred[0]
            if i==30:
                if(k==0):
#                     clf = joblib.load('model2.pkl')
#                     y_pred = clf.predict(cleaned_landmark)
#                     image = cv2.putText(image, str(y_pred[0]), (50,150), cv2.FONT_HERSHEY_SIMPLEX,  3, (0,0,255), 2, cv2.LINE_AA)
                    c=str(dictionary[y_pred[0]])
                    if(c in vowels):
                        k=(k-1)
                    if len(s)!=0 and c=="అం":
                        c="ం"
                    s+=c
                if(k==1):
#                     clf = joblib.load('model2.pkl')
#                     y_pred = clf.predict(cleaned_landmark)
#                     image = cv2.putText(image, str(y_pred[0]), (50,150), cv2.FONT_HERSHEY_SIMPLEX,  3, (0,0,255), 2, cv2.LINE_AA) 
                    last=s[-1]
                    c=last
                    s=str(s[:-1]+str(fun1(last, dictionary[str(y_pred[0])])))
                if(k==2):
#                     clf = joblib.load('model1.pkl')
#                     y_pred = clf.predict(cleaned_landmark)
#                     image = cv2.putText(image, str(y_pred[0]), (50,150), cv2.FONT_HERSHEY_SIMPLEX,  3, (0,0,255), 2, cv2.LINE_AA) 
                    if s[-1]==c:
                        last=s[-1]
#                         print(y_pred[0])
                        s=s[:-1]+fun2(last,dictionary[str(y_pred[0])],c)
                    else:
                        last=s[len(s)-3:]
                        s=s[:len(s)-3]+fun2(last,dictionary[str(y_pred[0])],c)
#                     print(last,dictionary[str(y_pred[0])],c)
                print(s)
                k=(k+1)%3
                i=0
                if k==0:
                    print("అక్షరాని చూపెటండి: ")
                if k==1:
                    print("వత్తు చూపెటండి: ")
                if k==2:
                    print("గునితాని చూపెటండి: ")
            if current==y_pred[0]:
                i+=1
            else:
                current=y_pred[0]
                i=0
        image = cv2.putText(image, str(y_pred[0]), (1000,150), cv2.FONT_HERSHEY_SIMPLEX,  3, (0,0,255), 2, cv2.LINE_AA) 
    cv2.imshow('MediaPipe Hands', image)
    if key==ord(' '):
        s+=" "
        print("స్పెస్: ")
        print(s)
        k=0
    if key==ord('s'):
        mytext = s
        language = 'te'
        myobj = gTTS(text=mytext, lang=language, slow=False)
        myobj.save("sound.mp3")
        os.system("play sound.mp3")
#         break

hands.close()
cap.release()
cv2.destroyAllWindows()
```
